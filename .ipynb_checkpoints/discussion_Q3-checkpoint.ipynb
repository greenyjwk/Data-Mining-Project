```python
import pandas as pd
import numpy as np
import csv

filename = "soybean.csv"
df = pd.read_csv(filename)

# Question 1
print("Q1) The number of instance is : ",  df.shape[0] )

# Question 2
print("Q2) The number of attribute is : ", df.shape[1] )
crop_hist = df["crop-hist"]

# Question 3
check = set()
for element in crop_hist:
    check.add(element)
    
check.remove('?') # Data Cleaning
print("Q3) How many possible values are there for the crop-hist attribute? : ", len(check) )

# Question 4
fungicide_contained = []
for column in df.columns:
    check = df[column].isin(["fungicide"])
    if check.any(): fungicide_contained.append(column)

print("Q4) Which of these attributes has fungcide as a possible value?",  fungicide_contained)
```
